# Introducción

El objetivo de esta investigación es desarrollar un modelo teórico sobre la dinámica de una gota de metal líquido cayendo debido a la fuerza de gravedad a través de un medio de distinta densidad, sometiéndola a un campo magnético homogéneo perpendicular con respecto al movimiento de la gota.  
    
Se tiene la hipótesis de que el campo magnético aplicado a la gota causaría un frenado magnético debido a corrientes parásitas inducidas en ella. También se cree que podría existir una deformación en la gota debido a estos mismo efectos electromagnéticos.  



# Formulación del problema

Para resolver este problema y tener una manera más controlada de manipular el movimiento de la gota, se realizaron experimentos de tal forma que la gota tuviera geometría similar a un disco encerrado entre dos paredes. Con esto en mente, el desarrollo teórico se hará de tal forma que la geometría de la gota coincida con la experimental.  
    Para empezar, se hará un balance de las fuerzas que afectan a la gota en su descenso, y así encontrar una expresión para la velocidad a la que se desplaza la gota.

## Balance de fuerzas

Como sabemos por las leyes de newton, un objeto se desplaza si sobre él actua una fuerza externa. Para este caso, a nuestra gota de metal líquido lo afectan varias fuerzas las cuales, al superponer dan resultado su descenso. Dicha dinámica se representa de la siguiente manera,  

\begin{equation}
m\frac{du}{dt}=F_{B}+F_{D}+F_{b}-F_{w}
\end{equation}
  
donde m es la masa de la gota, $u$ es la velocidad dependiente del tiempo de la gota cayendo, $t$ es la variable que representa al tiempo, $F_{B}$ es la fuerza magnética inducida sobre la gota, $F_{D}$ es la fuerza de arrastre, $F_{b}$ es la fuerza de flotación y $F_{w}$ es el peso de la gota.  
   
Como se verá más adelante, todas las expresiones de la fuerza apuntan sobre el eje $z$, por esa razón la ecuación (1) es una ecuación unidimensional, siendo el vector de gravedad el que apunta hacia los valores negativos de $z$. El lado izquierdo de la ecuación nos da la fuerza resultante por la que se desplaza la gota, mientras que del lado derecho se encuentran las fuerzas externas que actuan sobre la gota. La expresión para el peso, se encuentra con signo negativo ya que, para la configuración elegida en el experimento, ésta se encuentra en contra posición al resto.  
    
Se presentarán entonces las expresiones para cada una de las fuerzas presentes en la ecuación (1) a excepción de la fuerza magnética, su análisis se dejará posterior a presentar los resultados experimentales para poder explicarlos de mejor manera.
    



## Peso

\begin{equation}
F_{w} = m g.
\end{equation}
    
Donde g es la constante de gravedad igual a $9.8 \frac{m}{s^{2}}$


## Fuerza de flotación

\begin{equation}
F_{b} = \rho_{m} V g.
\end{equation}

    
Donde $\rho_{m}$ es la densidad del medio material por el cual cae la gota, ya sea un líquido o un gas y V es el volumen de la gota en cuestion.

## Fuerza de arrastre

\begin{equation}
 F_{D} = \frac{1}{2} C_{D} \rho_{m} A u^{2}
\end{equation}
    
Donde $C_{D}$ es el coeficiente de arrastre el cual experimenta la gota al moverse a través de un medio material y $A$ es el área frontal la cual impacta con el medio al momento del desplazamiento.


### Coeficiente de arrastre


En Mecánica de fluidos, el coeficiente de resistencia es una cantidad adimensional que se usa para cuantificar la resistencia de un objeto en un medio fluido. Para esta investigación se utilizó el coeficiente de fricción Bagueri - Bonadonna el cual se expresa de la siguiente manera
    
\begin{equation}
C_{D} = \frac{24 k_{s}}{Re} \Big{[}1 + 0.125 \Big{(}\frac{Re K_{N}}{k_{s}}\Big{)}^{\frac{2}{3}}\Big{]} + \frac{0.46 k_{n}}{1 + \frac{5330 k_{s}}{Re k_{N}}} .
\end{equation}
    
Donde

$ k_{N} = 10^{\alpha_{2} [-log(F_{N})]^{\beta_{2}}},$  

$ k_{s} = \frac{F_{s}^{\frac{1}{3}} + F_{s}^{\frac{-1}{3}}}{2},$

$ \alpha_{2} = 0.45 + \frac{10}{e^{2.5 log\rho_{r}^{-1}} + 30}, $

$ \beta_{2} = 1 - \frac{37}{e^{3 log \rho_{r}^{-1}} + 100}, $

$ F_{s} = f e^{1.3} \Big{(} \frac{d_{eq}^{3}}{LIS}\Big{)}, $

$ F_{N} = f^{2} e \Big{(}  \frac{d_{eq}^{3}}{LIS} \Big{)}, $

$ e = \frac{I}{L}, $

$ f = \frac{S}{I}, $
        
$ \rho_{r} = \frac{\rho_{m}}{\rho_{d}} $
    
Siendo $\rho_{r}$ la propoción de las densidades del medio y de la densidad de la gota de metal líquido.
Para este caso I y L tienen la misma magnitud, que corresponden al diámetro de la gota, mientras que S será la distancia en la que queda comprimida la gota.



## Propiedades geométricas de una gota entre placas

    
Ya que aproximar la geometría de la gota a la de un disco era lo que se pretendía idealmente, en la práctica la forma que toma la gota de metal líquido al colocarla entre dos placas paralelas no es precisamente la de un disco sino la de un ovoide, y ya que se necesita el área y el volumen de la gota para calcular la magnitud de la fuerza de arrastre, se hará el cálculo de estas magnitudes apartir de la expresión matemática para un ovoide la cual es
    
\begin{equation}
\Big{(}\frac{x}{A}\Big{)}^{n} + \Big{(}\frac{y}{B}\Big{)}^{m}=1,
\end{equation}
    
donde $n$ y $m$ son números reales que se ajustarán para aproximar mejor a la geometría de la gota, $A$ y $B$ son la longitud de los ejes en $x$ y $y$ correspondientemente. De hecho, $A$ y $B$ se pueden expresar en términos de I y S de la siguiente manera
    
$ A = \frac{I}{2} $  
    
$ B = \frac{S}{2} $
    
Ya que se quiere conocer la magnitud del área y del volumen de la gota de metal líquido, las variables $x$ y $y$ no tienen nada que ver con las coordenadas elegidas experimentalmente para el descenso de la gota, sólo se usarán para resolver las integrales que se plantearán a continuacción.

### Área de referencia de la gota

El área de referencia de la gota se puede obtener haciendo la integral con respecto a $x$, despejando a $y$ de la ecuación (6) y convirtiendola en una función de $x$. 
    
\begin{equation}
y(x) = B\Big{[}1 -  \Big{(}\frac{x}{A}\Big{)}^{n}\Big{]}^{\frac{1}{m}}
\end{equation}
        
Se escogen los límites de integración de 0 a A, para así no tener resultados negativos en el radical, pero se debe multiplicar a la integral por 4, así
    
$ A = 4\int_{0}^{A} y(x) dx. $

Resolviendo la integral, obtenemos una expresión para el área de referencia la cual es
    
\begin{equation}
A = \frac{I S}{n} \beta \Big{(}\frac{1}{n},\frac{1}{m} +1\Big{)},
\end{equation}
    
donde $\beta$ es la función beta definida como
    
$ \beta(x,y)=\frac{\Gamma(x) \Gamma (y)}{\Gamma(x + y)}, $
    
y $\Gamma$ es la función gamma definida como 
    
$ \Gamma(x) = (x -1)!$
    
Un punto interesante es que para esta función gamma, si n >> 1 y m >> 1  $\Longrightarrow \beta \Big{(}\frac{1}{n},\frac{1}{m} +1\Big{)} \approx n $. Entonces el área de referencia de la gota quedaría como $A = IS$. Siendo ésta el área de un rectángulo de base I y altura S, lo que sería igual a que tuvieramos la figura de un disco en lugar de un ovoide.

### Volumen de la gota


Para encontrar el volumen de la gota a partir de la expresión matemática de un ovoide, se recurre a encontrar el volumen de un sólido de revolución a partir de la curva que se crea al despejar $y$ de dicha expresión.
    
$ V = 4 \pi \int_{0}^{\frac{I}{2}} |y(x)|^2 dx. $
    
Los límites de integración se escogen sólo en la parte positiva de $x$ para no tener valores negativos dentro del radical, y se multiplica a la integral por dos para obtener así el volumen completo del sólido de revolución. Resolviendo la integral nos queda

\begin{equation}
V= \frac{\pi S I^{2}}{2n} \beta \big{(} \frac{2}{n} , \frac{1}{m} + 1 \big{)}.
\end{equation}
    
Al igual que en el caso del área, resulta interesante notar que si n >> 1 y m >> 1  $ \Longrightarrow \beta \Big{(}\frac{2}{n},\frac{1}{m} +1\Big{)} \approx \frac{n}{2}$, y al sustituir I por 2r, donde r es el radio del círculo, entonces el Volumen de la gota resulta ser el volumen de un disco siendo $ V = \pi r^{2} S $.

```python

A=2 #eje horizontal
B=1 #eje vertical


#definiendo la función y(x)
def y(x):
    return B*((1 - (x/A)**n)**(1/m))
x= np.linspace(0,A,100)


fig, axes = plt.subplots(1,2,sharey = True, figsize = (6*A, 4*B))
fig.suptitle("Área de referencia de la gota", fontsize = 18)


#figura izquierda
n=4 
m=2 
axes[0].plot(x,y(x),"gray")
axes[0].plot(-x,y(x),"gray")
axes[0].plot(x,-y(x),"gray")
axes[0].plot(-x,-y(x),"gray")
axes[0].set_aspect('equal')
axes[0].set_title("n = {}  y  m = {}".format(n,m), alpha = 0.7, fontsize = 14)
axes[0].set_yticks([-1,0,1])
axes[0].set_xticks([-2,-1,0,1,2])
axes[0].set_ylim(-1.3, 1.3)
axes[0].set_xlim(-2.3, 2.3)
axes[0].fill_between(x,y(x),-y(x), color = "#C9C9C9" )
axes[0].fill_between(-x,y(x),-y(x),color = "#C9C9C9" )


#figura derecha
n=15
m=15
axes[1].plot(x,y(x),"gray")
axes[1].plot(-x,y(x),"gray")
axes[1].plot(x,-y(x),"gray")
axes[1].plot(-x,-y(x),"gray")
axes[1].set_aspect('equal')
axes[1].set_title("n = {}  y  m = {}".format(n,m), alpha = 0.7, fontsize = 14)
axes[1].set_xticks([-2,-1,0,1,2])
axes[1].set_xlim(-2.3, 2.3)
axes[1].fill_between(x,y(x),-y(x), color = "#C9C9C9" )
axes[1].fill_between(-x,y(x),-y(x),color = "#C9C9C9" )

plt.show()
```

# Modelo analítico

Ahora hemos de buscar una solución análitica para así poder contrastarla con los resultados esperimentales que más adelante describiremos.
    
Teniendo ya definidas las fuerzas que actuan en este problema, podemos sustituir dichas expresiones en la ecuación (1) y así encontrar una expresión para la velocidad que describe el comportamiento de una gota de metal líquido cayendo entre dos placas a través de un medio material.
    
\begin{equation}
m\frac{du(t)}{dt}= \frac{1}{2} C_{D} \rho_{m} A u^{2}(t) + \rho_{m} V g - mg
\end{equation}
    
Hay que recordar que la expresión para la fuerza magnética se omitirá, hayando así una solución únicamente para la parte hidrodinámica. En la sección de discusión se discutirá el porqué.

## Solución analítica para la parte hidrodinámica

Resolviendo la ecuación diferencal (10) obtenemos 
    
\begin{equation}    
u(t) = \left\lbrace
\begin{array}{ll}
si \hspace{0.3cm} u_{0} < \big{|}\frac{c_{2}}{c_{1}} \big{|} &  \frac{c_{2}}{c_{1}} tgh \Big{[} - c_{1} c_{2} t + ln\big{(} \frac{c_{2} + c_{1} u_{0}}{c_{2} - c_{1} u_{0}}\big{)} \Big{)}\Big{]}\\
si \hspace{0.3cm} u_{0} > \big{|}\frac{c_{2}}{c_{1}} \big{|} & \frac{c_{2}}{c_{1}} coth \Big{[} - c_{1} c_{2} t + ln\big{(} \frac{c_{1} u_{0} + c_{2}}{c_{1} u_{0} - c_{2}}\big{)} \Big{)}\Big{]}
\end{array}
\right.
\end{equation}

    
Siendo $u_{0}$ la condición inicial,y donde $c_{1}$ y $c_{2}$ son  
$$c_{1} = \sqrt{\frac{C_{D} \rho_{r}  \beta \big{(} \frac{1}{n},\frac{1}{m} +1\big{)}}{\pi I \beta\big{(} \frac{2}{n} , \frac{1}{m} + 1 \big{)}}}$$
    
$$ c_{2} = \sqrt{g (1 - \rho_{r})}$$

```python
#definiendo una función para la velocidad u(t)
def velocidad(t):
    if abs(u0)<c2/c1:
        return (c2/c1)*np.tanh(-c1*c2*t + np.arctanh(c1/c2*u0))
        
    elif abs(u0)>c2/c1:
        return c2/(c1*np.tanh(-c1*c2*t + np.arctanh(c2/(c1*u0))))
    else:
        return np.full(len(t),c2/c1)

t = np.linspace(0,1,50)
c1 = 2
c2 = 4
u0 = 0

plt.plot(t,velocidad(t),color = "#68CAE4", label="u(t)",linewidth = 3 )
plt.xlabel("s")
plt.ylabel("m/s")
plt.title("Gráfica para la velocidad analítica")
plt.legend()
plt.show()
```

## Contrastando gráfica experimental y teórica
